## Creating DataFrame

### using list

```python
student_data= [
    [60, 30, 55],
    [77, 45, 67],
    [23, 34, 66],
    [11, 23, 45]
]

pd.DataFrame(student_data, columns=['lab', 'theory', 'quize'])
```

### using Dict

```python
student_data= {
   'lab': [60, 77, 23, 11],
    'theory':[30, 45, 34, 23],
    'quize':[55, 67, 66, 45]
}

pd.DataFrame(student_data)
```
